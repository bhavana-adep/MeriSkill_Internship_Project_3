```python
#importing packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#reading the dataset

df=pd.read_csv("/content/drive/MyDrive/HR-Employee-Attrition.csv")
df.head()
```

```python
#droping Uneccesary columns

df.drop(['EmployeeCount','EmployeeNumber','StandardHours'],axis=1,inplace=True)
```

```python
#checking null values are presernt or not

df.isnull().sum()
```

```python
#NAN values are checking

df.isna().sum()
```

```python
#checking if any duplicate values are present or not

df.duplicated().sum()
```

```python
#rename the column name of TotalWorkingYears to Experience

df.rename(columns={'TotalWorkingYears':'Experience'},inplace=True)
```

```python
plt.figure(figsize=(12,6))
a=df.corr()
sns.heatmap(a,annot=True,cmap='coolwarm',fmt='.2f')
plt.title("Correlation Heatmap for Numerical Columns", fontsize=15)
```

```python
attr=df['Attrition'].value_counts().reset_index()
plt.figure(figsize=(7,5))
plt.title("Total Attrition",fontsize=15)
plt.pie(attr['Attrition'],labels=attr['index'],autopct='%1.2f%%')
plt.show()
```

Total Attrition percent are 16.12%

```python
attr=df.groupby(['Attrition','Gender']).size().reset_index()
attr.rename(columns={0:'Count'},inplace=True)
atr=attr[attr['Attrition']=='Yes']
plt.figure(figsize=(6,5))
plt.title("Attrition Vs Gender")
plt.ylabel("Gender")
plt.xlabel("Attrition")
plt.barh(atr['Gender'],atr['Count'],height=0.5,color='#339966')
plt.show()
```

In this graph represents attrition are more done by Male as compared to Female

```python
age=df.groupby(['OverTime','Age']).size().reset_index()
age.rename(columns={0:'Count'},inplace=True)
age=age[age['OverTime']=='Yes']
age=age.sort_values(by='Count',ascending=False)
plt.figure(figsize=(6,5))
plt.title("OverTime Vs Age")
plt.ylabel("OverTime")
plt.xlabel("Age")
plt.scatter(age['Age'],age['Count'])
plt.show()
```

The graph shows the age between 30 to 40 are doing more overtiming as compared to other age groups

```python
dis=df.groupby(['Attrition','DistanceFromHome']).size().reset_index()
dis.rename(columns={0:'Count'},inplace=True)
dis=dis[dis['Attrition']=='Yes']
a=dis[dis['DistanceFromHome']>15]
c=a['Count'].sum()
b=dis[dis['DistanceFromHome']<=15]
d=b['Count'].sum()
e=pd.DataFrame([c,d],columns=['Count'])
x=['>15','<=15']
plt.figure(figsize=(6,5))
plt.title("Distance from Home Vs Attrition Count")
plt.ylabel("Count")
plt.xlabel("Distance from Home")
plt.bar(x,e['Count'],width=0.4,color='#ff4d4d')
plt.show()
```

This graph represents distance from home in Km with respect to attrition distance from home is not the reason of attrition it may be other reasons

```python
per=df.groupby('PerformanceRating')['PercentSalaryHike'].mean().reset_index()
a=list(per['PerformanceRating'].unique())
plt.figure(figsize=(7,5))
plt.title("Average SalaryHike According to Performance Ratings",fontsize=15)
plt.pie(per['PercentSalaryHike'],labels=a,autopct='%1.2f%%')
plt.show()
```

The graph shows that for employees with a good performance rating, the salary hike is higher.

```python
numerical_col=['DailyRate','HourlyRate','MonthlyIncome','MonthlyRate']
plt.figure(figsize=(10,6))
for i, column in enumerate(numerical_col,1):
    plt.subplot(2,2,i)
    plt.boxplot(df[column])
    plt.xlabel(column)
```

In this graph MonthlyIncome have few outliers

```python
dep=df.groupby(['Department','JobRole','Attrition']).size().reset_index()
dep.rename(columns={0:'Count'},inplace=True)
dep_yes=dep[dep['Attrition']=='Yes']
plt.figure(figsize=(9,6))
plt.xticks(fontsize=5)
plt.xlabel("JobRole",fontsize=6)
plt.title("JobRole wise Attrition")
plt.bar(dep_yes['JobRole'],dep_yes['Count'],width=0.9,color='#993366')
plt.show()
```

In this graph Laboratory Technician are doing more attrition with respect to other jobroles

```python
martial_status=df.groupby(['MaritalStatus','Attrition']).size().reset_index()
martial_status.rename(columns={0:'Count'},inplace=True)
a=martial_status[martial_status['Attrition']=='Yes']
b=martial_status[martial_status['Attrition']=='No']
fig, axs=plt.subplots(1,2,figsize=(10,5))
axs[0].bar(a['MaritalStatus'],a['Count'])
axs[0].set_title("Marital Status wise Attrition_Yes",fontsize=15)
axs[0].set_xlabel("Marital Status",fontsize=13)
axs[0].set_ylabel("Count",fontsize=13)
axs[1].set_title("Marital Status wise Attrition_No",fontsize=15)
axs[1].set_xlabel("Marital Status",fontsize=13)
axs[1].set_ylabel("Count",fontsize=13)
axs[1].bar(b['MaritalStatus'],b['Count'],color='#339966')
plt.show()
```

This graphs illustrate that single people are more doing attrition with respect to other martial status

```python

plt.figure(figsize=(10,6))
plt.title("MonthlyIncome Distribution",fontsize=15)
plt.hist(df['MonthlyIncome'],bins=50,edgecolor='red',alpha=0.4)
plt.show()
```

The Most common range of Monthly Income are 2000 to 2500

```python
travel=df.groupby(['BusinessTravel','Gender','Attrition']).size().reset_index()
travel.rename(columns={0:'Count'},inplace=True)
travel=travel[travel['Attrition']=='Yes']
a=travel[travel['Gender']=='Female']
b=travel[travel['Gender']=='Male']
plt.figure(figsize=(7,5))
plt.title("Attrition with Respect to BusinessTravel and Gender",fontsize=15)
plt.xlabel("Count",fontsize=13)
plt.ylabel("BusinessTravel",fontsize=13)
plt.barh(a['BusinessTravel'],a['Count'],label='Female',height=0.5)
plt.barh(a['BusinessTravel'],b['Count'],left=a['Count'],label='Male',height=0.5,color='#993366')
plt.legend()
plt.show()
```

```python
avg_salary=df.groupby('EducationField')['MonthlyIncome'].mean().reset_index()
avg_salary=avg_salary.sort_values(by='MonthlyIncome',ascending=False)
plt.figure(figsize=(7,5))
exp=(0.08,0,0,0,0,0)
plt.title("Average MonthlyIncome with respect to EducationField",fontsize=15)
plt.pie(avg_salary['MonthlyIncome'],labels=avg_salary['EducationField'],autopct='%1.2f%%',explode=exp)
plt.show()
```

This graph shows relationship with MonthlyIncome and EducationField.The average income comes from Marketing EducationField as comapred to other EducationField

```python
env_sat=df.groupby(['EnvironmentSatisfaction','Attrition']).size().reset_index()
env_sat.rename(columns={0:'Count'},inplace=True)
env_sat=env_sat[env_sat['Attrition']=='Yes']
a=['Low','Medium','High','Very High']
plt.figure(figsize=(7,5))
plt.title("EnvironmentSatisfaction Vs Attrition",fontsize=15)
plt.xlabel("EnvironmentSatisfaction",fontsize=13)
plt.ylabel("Count",fontsize=13)
plt.bar(a,env_sat['Count'],width=0.4,color='#ff4d4d')
plt.show()
```

This graph shows relationship between Environment Satisfaction and Attrition when the employee are not comfortable with working environment they are more likely to leave their job

```python
exp=df.groupby(['YearsAtCompany','Attrition']).size().reset_index()
exp.rename(columns={0:'Count'},inplace=True)
exp=exp[exp['Attrition']=='Yes']
plt.figure(figsize=(7,5))
plt.title("YearsAtComapny Vs Attrition")
plt.xlabel("YearsAtComapny")
plt.bar(exp['YearsAtCompany'],exp['Count'],color='orange')
plt.show()
```

This graph represents the experience between 1 to 5 years are doing more attrition
